```python
import pandas as pd
```

```python
df=pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
df.info()
```

```python
df.columns=['CustomerID', 'Gender', 'Age', 'AnnualIncome', 'SpendingScore']
```

```python
df['Age_Group']=pd.cut(
    df['Age'],
    bins=[0,20,35,60,100],
    labels=['Teen', 'Young Adult', 'Adult', 'Senior']
)
```

```python
df['Income_Level']=pd.cut(
    df['AnnualIncome'],
    bins=[0,40,70,150],
    labels=['Low', 'Medium', 'High']
)
```

```python
df['Spending_Type']=df['SpendingScore'].apply(
    lambda x: 'Low Spender' if x<40 else 'Medium spender' if x<70 else 'High Spender'
)
```

```python
df['Gender_Num']=df['Gender'].map({'Male': 0, 'Female': 1})
```
